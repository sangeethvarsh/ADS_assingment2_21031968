## A mini World Bank API Example

To install World Bank API, execute this line in the command shell: 

```
pip install wbgapi
```

To learn more about the World Bank API, click [here](https://pypi.org/project/wbgapi/).

````python
import pandas as pd
import wbgapi as wb
````

Choose countries:

````python
wb.source.info()
````

````python
country_codes =["ABW","AFW","BGD","ASM","BLR","CEB","CPV","DNK","DOM","HKG"] # United Kingdom and France
````

Search for identifiers:

````python
wb.search("economy")
````

We are interested in:

- GDP (current US$)

- Population, total

````python
indicator_ids = ["SP.URB.TOTL.IN.ZS","NV.AGR.TOTL.ZS", "EG.ELC.ACCS.ZS","AG.LND.IRIG.AG.ZS","SP.URB.TOTL.IN.ZS","SP.URB.TOTL"]
````

Display identifiers info:

````python
wb.series.info(indicator_ids)
````

Get the data:

````python
# specicif time range
data = wb.data.DataFrame(indicator_ids, country_codes, time=range(2016, 2021))
data
````

````python
# most recent years
data_1 =  wb.data.DataFrame(indicator_ids, country_codes, mrv=5)
data_1
````

````python
# Statistical Description of the retrieved dataset
data_1.describe().T
````

````python
wb.data.DataFrame('EG.ELC.ACCS.ZS', ["ABW","AFW","BGD","ASM","BLR","CEB","CPV","DNK","DOM","HKG"], time=range(2016,2020),numericTimeKeys=True).transpose().plot()
````

````python
wb.data.DataFrame('NV.AGR.TOTL.ZS', ["ABW","AFW","BGD","ASM","BLR","CEB","CPV","DNK","DOM","HKG"], time=range(2016,2020),numericTimeKeys=True).transpose().plot()
````

````python
wb.data.DataFrame('SP.URB.TOTL.IN.ZS', ["ABW","AFW","BGD","ASM","BLR","CEB","CPV","DNK","DOM","HKG"], time=range(2016,2020),numericTimeKeys=True).transpose().plot()
````

````python
# Creating the series dataframe
data1_c = data_1.groupby('series')
data1_c.describe()
````
